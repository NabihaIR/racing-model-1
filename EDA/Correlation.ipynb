```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('data.csv')

if 'SECTION_TIME_MS' in data.columns:
    # Summary statistics for SECTION_TIME_MS
    section_time_stats = data['SECTION_TIME_MS'].describe()

    # Plot distribution for SECTION_TIME_MS
    plt.figure(figsize=(10, 5))
    sns.histplot(data['SECTION_TIME_MS'], kde=True, bins=30, color='purple')
    plt.title('SECTION_TIME_MS Distribution')
    plt.xlabel('SECTION_TIME_MS (ms)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
else:
    print("The column 'SECTION_TIME_MS' is not found in the dataset.")
```
