```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data for the right and left tracks
right_track = pd.read_excel('right_track.xlsx')
left_track = pd.read_excel('left_track.xlsx')

# Filter data to include only points where -200 <= WORLDPOSY <= 300
right_track_filtered = right_track[(right_track['WORLDPOSY'] >= -200) & (right_track['WORLDPOSY'] <= 300)]
left_track_filtered = left_track[(left_track['WORLDPOSY'] >= -200) & (left_track['WORLDPOSY'] <= 300)]
 
# Plot the filtered left and right track points on a 2D grid
plt.figure(figsize=(12, 8))
plt.scatter(right_track_filtered['WORLDPOSX'], right_track_filtered['WORLDPOSY'], c='orange', label='Right Track', s=10)
plt.scatter(left_track_filtered['WORLDPOSX'], left_track_filtered['WORLDPOSY'], c='teal', label='Left Track', s=10)

# Set labels and title
plt.xlabel('WORLDPOSX')
plt.ylabel('WORLDPOSY')
plt.title('Left and Right Track Visualization (Filtered Y Values)')
plt.legend()
plt.grid(True)
plt.show()
```
