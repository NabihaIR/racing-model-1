```python
import pandas as pd

# Load CSV file
file_path = '../data.csv'

# List of columns 
columns_to_select = [
    'FB_BRAKE', 'MB_BRAKE', 'FT_THROTTLE', 'MT_THROTTLE', 'FS_STEERING', 'FS_BRAKE', 'MS_STEERING', 'MS_BRAKE', 'MS_THROTTLE',
    'FB_SPEED', 'FT_SPEED', 'MT_SPEED', 'FS_SPEED', 'MS_SPEED',
    'FS_DIST_FROM_LEFT', 'FT_DIST_FROM_LEFT', 'MS_DIST_FROM_LEFT', 'MT_DIST_FROM_LEFT', '360_DIST_FROM_LEFT',
    '360_ANGLE_APEX_1', '360_SPEED', '468_DIST_FROM_LEFT', '468_ANGLE_APEX_2', '468_SPEED',
    'FB_LAP_DIST', 'MB_LAP_DIST', 'FS_LAP_DIST', 'MS_LAP_DIST', 'FT_LAP_DIST', 'MT_LAP_DIST',
    'FB_DIST_APEX_1', 'FS_DIST_APEX_1', 'FT_DIST_APEX_1', 'MS_DIST_APEX_1', 'MS_DIST_APEX_2', 
    'MB_DIST_APEX_1', 'MT_DIST_APEX_1', 'MT_DIST_APEX_2', 'SECTION_TIME_MS'
]

# Load only the specified columns from the CSV file
data = pd.read_csv(file_path, usecols=columns_to_select)

# Fixed apex distances
apex_1_dist = 387
apex_2_dist = 495

# Calculate LAP_DIST_FROM_APEX
data['FB_LAP_DIST_FROM_APEX_1'] = apex_1_dist - data['FB_DIST_APEX_1']
data['FS_LAP_DIST_FROM_APEX_1'] = apex_1_dist - data['FS_DIST_APEX_1']
data['FT_LAP_DIST_FROM_APEX_1'] = apex_1_dist - data['FT_DIST_APEX_1']
data['MS_LAP_DIST_FROM_APEX_1'] = apex_1_dist - data['MS_DIST_APEX_1']
data['MS_LAP_DIST_FROM_APEX_2'] = apex_2_dist - data['MS_DIST_APEX_2']
data['MB_LAP_DIST_FROM_APEX_1'] = apex_1_dist - data['MB_DIST_APEX_1']
data['MT_LAP_DIST_FROM_APEX_1'] = apex_1_dist - data['MT_DIST_APEX_1']
data['MT_LAP_DIST_FROM_APEX_2'] = apex_2_dist - data['MT_DIST_APEX_2']

# Drop orginial apex distance columns 
data = data.drop(columns=['FB_DIST_APEX_1', 'FS_DIST_APEX_1', 'FT_DIST_APEX_1', 'MS_DIST_APEX_1', 'MS_DIST_APEX_2', 
                          'MB_DIST_APEX_1', 'MT_DIST_APEX_1', 'MT_DIST_APEX_2'])

# Save the updated data to a new CSV file
data.to_csv('data_new.csv', index=False)

# Display the first few rows of the modified data (optional)
data.head()
```
